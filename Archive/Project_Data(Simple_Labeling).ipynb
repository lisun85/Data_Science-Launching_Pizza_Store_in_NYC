```python
import pandas as pd
```

```python
a = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_161015.txt")
b = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_161008.txt")
c = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_161001.txt")

d = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_160924.txt")
e = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_160917.txt")
f = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_160910.txt")
g = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_160903.txt")

h = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_160827.txt")
i = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_160820.txt")
j = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_160813.txt")
k = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_160806.txt")

l = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_160730.txt")
m = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_160723.txt")
n = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_160716.txt")
o = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_160709.txt")
```

```python
#Concat datasets for 15 weeks
df = pd.concat([o,n,m,l,k,j,i,h,g,f,e,d,c,b,a])

df
```

```python
#Data Cleaning - viewing all columns
df.columns
```

```python
#Declaring new variable and stripping all column names of spaces
new_col_names = [col.strip() for col in df.columns]
new_col_names
```

```python
#replace messy column names with stripped/clean column names
df.set_axis(labels = new_col_names, axis=1, inplace=True)
```

```python
df.columns
```

```python
#combining Entries and Exits to get total Foot_Traffic
df["Foot_Traffic"] = df.ENTRIES + df.EXITS
df
```

```python
#I noticed that data in Foot_Traffic is a cumulative number, so what we want is the incremental foottraffic
#using diff(), I extracted the incremental Foot_Traffic and called it Actual_FT

df["Actual_FT"] = df.Foot_Traffic.diff(1)
df.head(100)
```

```python
#Noticed negatives, which means, if stations switch, the cumulative numbers are not the same, so what we must do is delete the negatives
df.groupby("DATE").Actual_FT.mean()
```

```python
#took the negatives and replaced it with 0
df[df.Actual_FT < 0] = 0
```

```python
#these numbers look better, but 07/02 data looks a bit more extreme
df.groupby("DATE").Actual_FT.sum()
```

```python
#store the data into variable
stationData = df.groupby("STATION").Actual_FT.mean()
dateData = df.groupby("DATE").Actual_FT.sum()
dateData
```

```python
# The Stations scoped in this project are: Time's Square, Hearld Sq, Bryant Park, Grand Central, Pen Station, 14St, Astor Place
#'TIMES SQ-42 ST', '34 ST-HERALD SQ', '42 ST-BRYANT PK', 'GRD CNTRL-42 ST', '34 ST-PENN STA', '14 ST-UNION SQ', 'ASTOR PL'

df2 = df[(df["STATION"] == 'TIMES SQ-42 ST') |
        (df["STATION"] == '34 ST-HERALD SQ')|
        (df["STATION"] == '42 ST-BRYANT PK')|
        (df["STATION"] == 'GRD CNTRL-42 ST')|
        (df["STATION"] == '34 ST-PENN STA') |
        (df["STATION"] == '14 ST-UNION SQ') |
        (df["STATION"] == 'ASTOR PL') |
        (df["STATION"] == '34 ST-HUDSON YD') |
        (df["STATION"] == '23 ST') |
         (df["STATION"] == '28 ST') | 
         (df["STATION"] == '33 ST')]

stations = df2.groupby("STATION").Actual_FT.sum().sort_values(ascending = False)

stations
```

```python
#plotting a bar graph showing the foot_traffic of all stations in comparison to each other
plt.bar(stations.index, stations.values)
plt.xticks(rotation = 90)
```
