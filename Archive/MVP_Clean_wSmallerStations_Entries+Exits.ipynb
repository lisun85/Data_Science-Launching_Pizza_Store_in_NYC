```python
#CREATION OF PIZZA_DAILY
#maximum ENTRIES per day. Pizza_daily (newly created) is only ENTRIES, will create Exits later

pizza_daily = (df_pizza
                  .groupby(["C/A", "UNIT", "SCP", "STATION","DATE", "TIME"], as_index=False)
                  ['ENTRIES', 'EXITS'].last())

pizza_daily
```

```python
#Create two columns with shift(1)

pizza_daily[['PREV_DATE', 'PREV_ENTRIES', 'PREV_EXITS']] = (pizza_daily
                                                 .groupby(['C/A', 'UNIT', 'SCP', 'STATION'])['DATE', 'ENTRIES', 'EXITS']
                                                 .apply(lambda grp: grp.shift(1))
                                             )
pizza_daily.head()
```

```python
#fix with formula
def get_daily_counts(row, max_counter):
    counter = row['ENTRIES'] - row["PREV_ENTRIES"]
    if counter < 0:
        counter = -counter
    if counter > max_counter:
        print(row['ENTRIES'], row['PREV_ENTRIES'])
        return 0
    return counter
_ = pizza_daily.apply(get_daily_counts, axis=1, max_counter=1000000)
```

```python
#THIS IS THE FIX (ENTRIES)
def get_daily_counts(row, max_counter):
    counter = row['ENTRIES'] - row['PREV_ENTRIES']
    if counter < 0:
        counter = -counter
    if counter > max_counter:
        print(row['ENTRIES'], row['PREV_ENTRIES'])
        counter = min(row['ENTRIES'], row['PREV_ENTRIES'])
    if counter > max_counter:
        return 0
    return counter

pizza_daily['DAILY_ENTRIES'] = pizza_daily.apply(get_daily_counts, axis=1, max_counter=1000000)
```

```python
#THIS IS THE FIX (EXITS)
def get_daily_counts(row, max_counter):
    counter = row['EXITS'] - row['PREV_EXITS']
    if counter < 0:
        counter = -counter
    if counter > max_counter:
        print(row['EXITS'], row['PREV_EXITS'])
        counter = min(row['EXITS'], row['PREV_EXITS'])
    if counter > max_counter:
        return 0
    return counter

pizza_daily['DAILY_EXITS'] = pizza_daily.apply(get_daily_counts, axis=1, max_counter=1000000)
```

```python
#how many have this?
pizza_daily[pizza_daily.DAILY_EXITS > 1000000]
```

```python
#combining DAILY Entries and Exits to get total DAILY Total_Foot_Traffic 
pizza_daily["TOTAL_DAILY_FT"] = pizza_daily.DAILY_ENTRIES + pizza_daily.DAILY_EXITS
pizza_daily
```

```python
#Daily Averages
station_daily = pizza_daily.groupby(["STATION", "DATE"])['DAILY_ENTRIES','DAILY_EXITS', 'TOTAL_DAILY_FT'].mean().reset_index()
station_daily
```

```python
#STATION TOTAL (ALL TIMES)
station_total = pizza_daily.groupby(["STATION"]).sum().sort_values('TOTAL_DAILY_FT', ascending = False).reset_index()
station_total
```

```python
#SPLITTING OUT ONLY MIDNIGHT TO 5AM
pizza_daily_5am = pizza_daily[(pizza_daily["TIME"] == "00:00:00") |
                              (pizza_daily['TIME'] == '04:00:00') |
                              (pizza_daily['TIME'] == '01:00:00') |
                              (pizza_daily['TIME'] == '05:00:00')]

pizza_daily_5am
```

```python
station_average_5am = pizza_daily_5am.groupby(['STATION']).mean().sort_values('TOTAL_DAILY_FT', ascending = False).reset_index()

station_average_5am
```
