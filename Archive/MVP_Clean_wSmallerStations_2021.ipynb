```python
#CREATION OF PIZZA_DAILY
#maximum ENTRIES per day. Pizza_daily (newly created) is only ENTRIES, will create Exits later

pizza_daily = (df_pizza
                  .groupby(["C/A", "UNIT", "SCP", "STATION","DATE", "TIME"], as_index=False)
                  .ENTRIES.last())

pizza_daily
```

```python
#Create two columns with shift(1)

pizza_daily[['PREV_DATE', 'PREV_ENTRIES']] = (pizza_daily
                                                 .groupby(['C/A', 'UNIT', 'SCP', 'STATION'])['DATE', 'ENTRIES']
                                                 .apply(lambda grp: grp.shift(1))
                                             )
pizza_daily.head()
```

```python
#fix with formula
def get_daily_counts(row, max_counter):
    counter = row['ENTRIES'] - row["PREV_ENTRIES"]
    if counter < 0:
        counter = -counter
    if counter > max_counter:
        print(row['ENTRIES'], row['PREV_ENTRIES'])
        return 0
    return counter
_ = pizza_daily.apply(get_daily_counts, axis=1, max_counter=1000000)
```

```python
def get_daily_counts(row, max_counter):
    counter = row['ENTRIES'] - row['PREV_ENTRIES']
    if counter < 0:
        counter = -counter
    if counter > max_counter:
        print(row['ENTRIES'], row['PREV_ENTRIES'])
        counter = min(row['ENTRIES'], row['PREV_ENTRIES'])
    if counter > max_counter:
        return 0
    return counter

pizza_daily['DAILY_ENTRIES'] = pizza_daily.apply(get_daily_counts, axis=1, max_counter=1000000)
```
